Given a dictionary of words and a string made up of those words (no spaces), return the original sentence in a list. If there is more than one possible reconstruction, return any of them. If there is no possible reconstruction, then return null.

For example, given the set of words 'quick', 'brown', 'the', 'fox', and the string "thequickbrownfox", you should return ['the', 'quick', 'brown', 'fox'].

Given the set of words 'bed', 'bath', 'bedbath', 'and', 'beyond', and the string "bedbathandbeyond", return either ['bed', 'bath', 'and', 'beyond] or ['bedbath', 'and', 'beyond'].

```python
import numpy as np
import unittest


def zero_out(matrix: np.array) -> None:
    """Given an NxM matrix zero out all rows and columns that contain at least one zero."""

    
    zero_cols = matrix.all(axis=0)
    zero_rows = matrix.all(axis=1)
    matrix[:, zero_cols] = 0
    matrix[zero_rows, :] = 0
    print(zero_cols)
    print(zero_rows)


def test_zero_out(self):
    matrix = np.array([[5, 3, 2, 1],
                       [-3, 0, 5, 0],
                       [0, -1, 2, 6]])
    zeroed_out_matrix = np.array([[0, 0, 2, 0],
                                  [0, 0, 0, 0],
                                  [0, 0, 0, 0]])
    zero_out(matrix)
    np.testing.assert_array_equal(matrix, zeroed_out_matrix)
```

```python
matrix = np.array([[5, 3, 2, 1],
                       [-3, 0, 5, 0],
                       [0, -1, 2, 6]])

zero_out(matrix)
```

```python
def find_sentence(s, dictionary):
    starts = {0: ''}
    for i in range(len(s) + 1):
        print("i-",i)
        new_starts = starts.copy()
        print(new_starts)
        for start_index, _ in starts.items():
            print("si", start_index, _)
            word = s[start_index:i]
            print("word", word)
            if word in dictionary:
                new_starts[i] = word
                print("new starts[i]", new_starts[i])
        print("new_starts_new", new_starts)
        starts = new_starts.copy()
        print("starts", starts)

    result = []
    current_length = len(s)
    print(current_length)
    if current_length not in starts:
        return None
    while current_length > 0:
        word = starts[current_length]
        print(word)
        current_length -= len(word)
        result.append(word)

    return list(reversed(result))
```

```python
s = "bedbathandbeyond"
d = {"bed", "bath", "bedbath", "and", "beyond"}

find_sentence(s, d)
```
